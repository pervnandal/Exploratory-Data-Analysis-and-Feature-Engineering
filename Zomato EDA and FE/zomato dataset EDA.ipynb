```python
df_country = pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
df.columns
```

```python
final_df = pd.merge(df, df_country, on='Country Code', how = 'left') 
# on is used to match the column which is used to combine
# how is used to define what kind of join we want to perform - left, right, or any other
```

```python
# to check datatypes only
final_df.dtypes
```

```python
# final_df.Country.value_counts() # get to total count of each country
# final_df['Country'].value_counts().index # works same as line below
country_names = final_df.Country.value_counts().index # stores all names of country
```

```python
country_val = final_df.Country.value_counts().values # shows the value of each country respectively
```

```python
# Pie Chart
# plt.pie(country_val,labels=country_names) # but this looks very bad as data is not very well spread

# Pie Chart - top 3 countries that uses zomato
plt.pie(country_val[:3],labels=country_names[:3], autopct='%1.2f%%') 
# autopct is used to show percentage in pie chart, 1.2f means show 2 decimal values after a decimal 
```

Osbervation : **Zomato maximum records or transaction are form india**
*after that USA and then UK*

```python
final_df.columns
```

```python
# this will get the size of each rating 
# final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size()

# this is done to convert it into dataframe, including that size column, it's column name will be 0
# reset_index will reset the index we were orignally getting
# final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index()

# Now we need to rename that column
ratings = final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating count'})
```

***Obsevations***
1. when rating is between 4.5 to 4.9 ---> it is excellent 
2. when rating is between 4.0 to 4.4 ---> it is very good
3. when rating is between 3.5 to 3.9 ---> it is good
4. when rating is between 2.5 to 3.4 ---> it is average
5. when rating is between 2.0 to 2.4 ---> it is poor
6. Many people gave zero ratings

```python
ratings.head()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12,6) # this will increase figure size
sns.barplot(x="Aggregate rating",y="Rating count",data= ratings)
```

```python
# now lets color code this barplot acc. to rating color
# hue will color it but will not match colors correctly, so we will use pallete to match those colors 
sns.barplot(x="Aggregate rating",y="Rating count",data= ratings, hue="Rating color",palette=['blue','red','orange','yellow','green','green'])
```

***Observation***
1. Not rated count is very high
2. Maximum number of ratings is between 2.5 to 3.4 

from 2nd observation what we can do is take the average of it and use it to fill missing values

```python
# Count Plot - used for plotting categorical data
sns.countplot(x='Rating color', data= ratings, palette=['blue','red','orange','yellow','green','green'])
```
